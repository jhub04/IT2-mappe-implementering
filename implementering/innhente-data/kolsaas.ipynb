```python
fil = open("kolsaas.csv")
data = fil.read()
print(data)
```

```python
linjer = data.split("\n")
tur = []

for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
```

```python
import matplotlib.pyplot as plt

x = []
y = []


for e in tur[1:]:
    x.append(e[0])

for e in tur[1:]:
    y.append(e[3])

plt.plot(x, y)
plt.show()
```
