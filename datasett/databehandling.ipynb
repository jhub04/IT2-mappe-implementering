### Datasett i Python

> En stor samling av data kalles ofte for et datasett

```python
import matplotlib.pyplot as plt

fil = open("./internet-users.csv")
data = fil.read()
linjer = data.split("\n")

brukere = []
år = []

for linje in linjer[:-1]:
    splittet_linje = linje.split(",")
    if splittet_linje[1] == "Norway":
        brukere.append(float(splittet_linje[5]))
        år.append(int(splittet_linje[3]))

plt.plot(år, brukere)
```

```python
# Rydder csv filen for firkanta brackets ([])

import csv

input_file = "List_of_world_cities_by_population_density.csv"
output_file = "Cleaned_List_of_world_cities_by_population_density.csv"

with open(input_file, newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    
    with open(output_file, 'w', newline='') as output_csvfile:
        csv_writer = csv.writer(output_csvfile)

        for row in csv_reader:
            cleaned_row = [cell.replace('[', '').replace(']', '') for cell in row]
            csv_writer.writerow(cleaned_row)

print(f"Cleaned CSV file saved as: {output_file}")
```
